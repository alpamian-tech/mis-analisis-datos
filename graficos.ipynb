```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Datos reales de tu clase (puedes cambiar estos números)
notas = [55, 62, 65, 70, 71, 72, 75, 80, 88, 95]

# 2. Configuración estética profesional
sns.set_theme(style="darkgrid")
plt.figure(figsize=(8, 5))

# 3. Crear el histograma con "bins" específicos
ax = sns.histplot(notas, kde=True, color="purple", bins=5)

# 4. Personalizar etiquetas para que se vea "super bonito"
plt.title("Distribución de Notas del Examen", fontsize=14, color='darkblue')
plt.xlabel("Calificación", fontsize=12)
plt.ylabel("Número de Estudiantes", fontsize=12)

# 5. Guardar como una nueva imagen
plt.savefig("analisis_notas.png")
print("¡Gráfico de notas generado con éxito!")
```

```python


# 1. Creamos un diccionario con datos de ejemplo (Notas de Estadística)
datos = {
    'Estudiante': ['Ana', 'Luis', 'Marta', 'Juan', 'Sofía'],
    'Parcial 1': [85, 70, 95, 60, 88],
    'Parcial 2': [90, 75, 92, 65, 84],
    'Asistencia %': [100, 85, 98, 70, 92]
}

# 2. Convertimos los datos en un "DataFrame" (una tabla inteligente)
df = pd.DataFrame(datos)

# 3. Calculamos el promedio y lo agregamos como nueva columna
df['Promedio Final'] = df[['Parcial 1', 'Parcial 2']].mean(axis=1)

# 4. Le damos formato "bonito" (Solo funciona en Notebooks)
# Esto pintará de verde las notas más altas y de rojo las bajas
tabla_estilizada = df.style.background_gradient(cmap='Greens', subset=['Promedio Final'])\
                           .highlight_max(color='lightgreen', subset=['Asistencia %'])\
                           .format(precision=1)

# 5. MOSTRAR LA TABLA
tabla_estilizada
```
